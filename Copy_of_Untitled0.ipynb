```python
#merging the data frames
import pandas as pd
data1={'key':['k0','k1','k2','k3'],
       'name':['chunnu','jithu','vasanthi','kalyani'],
       'age':[23,34,25,26]}
data2={'key':['k0','k1','k2','k3'],'address':['nagpur','jaipur','manipur','jodhpur'],'qualification':['ms','msc','mba','bba']}
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,"\n\n",df1)
res=pd.merge(df,df1,on='key')
res
```

```python
#merging dataframe with multiple keys
import pandas as pd
data1={'key':['k0','k1','k2','k3'],
       'key1':['k0','k1','k0','k1'],
       'name':['chunnu','jithu','vasanthi','kalyani'],
       'age':[23,34,25,26]}
data2={'key':['k0','k1','k2','k3'],'key1':['k0','k0','k0','k0'],'address':['nagpur','jaipur','manipur','jodhpur'],'qualification':['ms','msc','mba','bba']}
df=pd.DataFrame(data1)
df1=pd.DataFrame(data2)
print(df,"\n\n",df1)
res=pd.merge(df,df1,on=['key','key1'])
res
```
